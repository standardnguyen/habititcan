```python
from trello import TrelloListMonitor
import os
from dotenv import load_dotenv
```

```python
load_dotenv()

TRELLO_AUTOPROCESSING_TRASH_LIST_ID = os.getenv("TRELLO_AUTOPROCESSING_TRASH_LIST_ID")
```

```python
monitor = TrelloListMonitor(TRELLO_AUTOPROCESSING_TRASH_LIST_ID)
```

```python
def handle_trello_changes(diff):
    """Handle all the business logic when Trello changes"""
    if diff['added']:
        for card in diff['added']:            
            card_id = card['id']
            monitor.delete_card(card_id)

```

```python
monitor.monitor(interval=1,callback=handle_trello_changes)
```
